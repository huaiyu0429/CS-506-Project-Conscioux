```python
import json
import pandas as pd
import matplotlib.pyplot as plt

tweets_data_path = 'datatest5.txt'

tweets_data = []
tweets_file = open(tweets_data_path, "r")
for line in tweets_file:
    try:
        tweet = json.loads(line)
        tweets_data.append(tweet)
    except:
        continue
#print (len(tweets_data))
```

```python
tweets=pd.DataFrame(tweets_data)
#print(tweets.iloc[0:5,])
```

```python
from pandas.core.frame import DataFrame

entities = tweets['entities']#['hashtags']
#print(type(entities))
hashtags = []
for i in range(len(entities)):
    hashtags.append(entities[i]['hashtags'])
#print(hashtags)

tag = []
for i in range(len(hashtags)):
    if len(hashtags[i]) != 0:
        num = []
        for hash in hashtags[i]:
            #print(type(hash['text']))
            num.append(hash['text'])
        tag.append(num)
    else:
        tag.append([])
        
user_info = tweets['user']
location = []
for i in range(len(user_info)):
    location.append(user_info[i]['location'])
    
place_info = tweets['place']
country_code = []
for i in range(len(place_info)):
    if place_info[i] == None:
        country_code.append('None')
    else:
        country_code.append(place_info[i]['country_code'])

coordinates_original = tweets['coordinates']
coordinates = []
for i in range(len(coordinates_original)):
    if coordinates_original[i] == None:
        coordinates.append('None')
    else:
        coordinates.append(coordinates_original[i]['coordinates'])

c = {"tags" : tag, 'coordinates': coordinates, 'time': tweets['created_at'],
     'fav_count':tweets['favorite_count'], 'review':tweets['text'], 'country_code':country_code} 
     #'country_code': country_code}#将列表a，b转换成字典
data = pd.DataFrame(c)#将字典转换成为数据框
#data 
```

```python
data1 = data[data.country_code=='US']
data2 = data1[data.coordinates != 'None']
data2 = data2.reset_index(drop=True)
```

```python
data2["lon"] = data2["coordinates"].apply(lambda a : a[0])
data2["lat"] = data2["coordinates"].apply(lambda a : a[1])
data2
```

```python
data2.to_csv('data.csv', sep=',')
```
